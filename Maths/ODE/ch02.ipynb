# Planar Linear System

## Solve ODE system

$$
\begin{cases}
x' = y \\
y' = -x
\end{cases}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# difine the system of ODEs
# x' = y, y' = -x
# This represents a simple harmonic oscillator
# or a circular motion in the phase plane.
def system(t, z):
    x, y = z
    return [y, -x]

# define the grid for the direction field
x_vals = np.linspace(-5, 5, 20)
y_vals = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x_vals, y_vals)
U = Y
V = -X
N = np.sqrt(U**2 + V**2)
U, V = U/N, V/N  # normalize the vectors

plt.figure(figsize=(8, 8))
plt.quiver(X, Y, U, V, color='gray')

# plot the direction field
initial_conditions = [
    [1, 0],
    [0, 2],
    [-2, 0],
    [0, -3],
    [3, 3]
]

t_span = [0, 20]
t_eval = np.linspace(t_span[0], t_span[1], 300)

for z0 in initial_conditions:
    sol = solve_ivp(system, t_span, z0, t_eval=t_eval)
    plt.plot(sol.y[0], sol.y[1], label=f'init: {z0}')

plt.xlabel('x')
plt.ylabel('y')
plt.title("Direction Field and Curve for x'=y, y'=-x")
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```
