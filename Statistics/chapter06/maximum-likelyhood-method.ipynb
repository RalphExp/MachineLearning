# Exercise 6.1

```python
import numpy
import pyreadr
import scipy.stats as stats

from matplotlib import pyplot as plt
```

```python
ex6111data = pyreadr.read_r('../Data/ex6111.rda')['ex6111']
# print(ex6111data)

plt.hist(ex6111data, bins=10, density=True)
theta = numpy.average(ex6111data) / 4

print("theta:", theta)

x = numpy.linspace(0, 100, 1000)
y = stats.gamma.pdf(x, a=4, scale=theta)
plt.plot(x, y) 
```
