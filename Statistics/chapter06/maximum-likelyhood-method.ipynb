# Exercise 6.1

```python
import numpy
import pyreadr
import scipy.stats as stats

from matplotlib import pyplot as plt
```

```python
ex6111data = pyreadr.read_r('../Data/ex6111.rda')['ex6111']
# print(ex6111data)

plt.hist(ex6111data, bins=10, density=True)
theta = numpy.average(ex6111data) / 4

x = numpy.linspace(0, 100, 1000)
y = stats.gamma.pdf(x, a=4, scale=theta)

plt.plot(x, y, label=f'θ = {theta}')
plt.legend()
plt.show()
```

# Exercise 6.3

```python
data = numpy.array([19, 15, 76, 23, 24, 66, 27, 12, 25, 7, 6, 16, 51, 26, 39])
plt.hist(data, bins=10, density=True)

x = numpy.linspace(0, 100, 1000)
y = stats.gamma.pdf(x, a=1, scale=40)
plt.plot(x, y, label='θ = 40')

y = stats.gamma.pdf(x, a=1, scale=50)
plt.plot(x, y, label='θ = 50')
plt.legend()
plt.show()

s50 = 2/50 * data.sum()
s40 = 2/40 * data.sum()
chi30_95 = stats.chi2.ppf(q=0.95, df=30)
chi30_5 = stats.chi2.ppf(q=0.05, df=30)
print(f's50 = {s50}, s40 = {s40}, chi30_5 = {chi30_5}, chi30_95 = {chi30_95}')
```

```python
data = numpy.array([27, 13, 21, 24, 22, 14, 17, 26, 14, 22, 21, 24 ,19, 25, 15, 25, 23, 16, 20, 19])
θ0 = 23
mean = data.mean()
n = len(data)
Fisher = 1/mean
Wald = n*Fisher*(mean-θ0)**2
print('chi-W =', Wald)

pvalue = 1-stats.chi2.cdf(Wald, df=1)
print('pvalue =', pvalue)
```
